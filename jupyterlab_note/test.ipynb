```python
import matplotlib.pyplot as plt
import sys


def setup_matplotlib_font():
    """配置matplotlib中文显示
    
    该函数会根据不同操作系统设置合适的字体，确保中文能够正确显示。
    支持Windows、macOS和Linux系统。

    Example:
        >>> from src.helper_utils.font_utils import setup_matplotlib_font
        >>> setup_matplotlib_font()  # 调用后，matplotlib将正确显示中文
    """
    if sys.platform.startswith("win"):
        plt.rcParams["font.family"] = ["SimHei", "Microsoft YaHei"]
    elif sys.platform.startswith("darwin"):
        plt.rcParams["font.family"] = ["Arial Unicode MS", "Heiti TC"]
    elif sys.platform.startswith("linux"):
        plt.rcParams["font.family"] = ["Droid Sans Fallback", "DejaVu Sans", "sans-serif"]
    plt.rcParams["axes.unicode_minus"] = False  # 正确显示负号



setup_matplotlib_font()
print("Matplotlib字体设置已完成，当前字体配置:")
print(f"- 字体族: {plt.rcParams['font.family']}")
print(f"- 负号显示: {plt.rcParams['axes.unicode_minus']}")
```

```python
import matplotlib.pyplot as plt
import numpy as np  # 用于生成示例数据

# 1. 准备数据
x = np.linspace(0, 10, 100)  # 生成0到10之间的100个均匀数据点
y = np.sin(x)  # 计算正弦值

# 2. 创建图表并绘图
plt.figure(figsize=(8, 4))  # 创建画布，设置大小(宽, 高)
plt.plot(x, y)  # 绘制折线图

# 3. 添加图表元素（标题、坐标轴标签等）
plt.title("正弦曲线")  # 标题
plt.xlabel("X轴")      # X轴标签
plt.ylabel("Y轴")      # Y轴标签
plt.grid(True)         # 显示网格线

# 4. 显示图表
plt.show()
```

```python
x = np.linspace(0, 10, 20)
y1 = np.sin(x)
y2 = np.cos(x)

# 自定义线条颜色、样式、宽度；标记形状、大小
plt.plot(x, y1, color='red', linestyle='--', linewidth=2,
         marker='o', markersize=5, label='sin(x)')
plt.plot(x, y2, color='blue', linestyle='-', linewidth=2,
         marker='s', markersize=5, label='cos(x)')

plt.legend()  # 显示图例（需在plot中设置label）
plt.title("自定义线条样式")
plt.show()
```

```python
plt.style.use('seaborn-v0_8-bright')  # 使用seaborn风格

x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.title("使用seaborn风格")
plt.show()
```

```python
# 创建2行2列的子图，当前激活第1个（行优先编号）
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("sin(x)")

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title("cos(x)")

plt.subplot(2, 2, 3)
plt.plot(x, x)
plt.title("y=x")

plt.subplot(2, 2, 4)
plt.plot(x, x**2)
plt.title("y=x²")

plt.tight_layout()  # 自动调整子图间距，避免重叠
plt.show()
```

```python
# 创建2行1列的子图，返回画布(fig)和子图数组(axes)
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

axes[0].plot(x, np.sin(x), color='red')  # 第1个子图
axes[0].set_title("sin(x)")  # 子图标题（注意用set_title而非title）

axes[1].plot(x, np.cos(x), color='blue')  # 第2个子图
axes[1].set_title("cos(x)")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 创建3D图形
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 生成数据
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# 绘制3D表面图
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# 添加颜色条
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_title('3D 表面图示例')
ax.set_xlabel('X轴')
ax.set_ylabel('Y轴')
ax.set_zlabel('Z轴')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 生成模拟数据
np.random.seed(42)  # 设置随机种子，确保结果可重现
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
sales = np.random.normal(100, 20, len(dates))
sales = np.cumsum(sales) + 1000  # 累积销售额，模拟增长趋势
expenses = sales * np.random.uniform(0.6, 0.8, len(dates))

# 创建一个包含多个子图的图表
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15))

# 第一个子图：销售额和支出趋势
ax1.plot(dates, sales, label='销售额', color='blue')
ax1.plot(dates, expenses, label='支出', color='red')
ax1.set_title('2023年销售和支出趋势')
ax1.set_ylabel('金额 (元)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 第二个子图：月度销售额柱状图
monthly_sales = sales.reshape(-1, 30).sum(axis=1)[:12]  # 简化的月度求和
months = ['一月', '二月', '三月', '四月', '五月', '六月',
          '七月', '八月', '九月', '十月', '十一月', '十二月']
ax2.bar(months, monthly_sales, color='green')
ax2.set_title('2023年月度销售额')
ax2.set_ylabel('金额 (元)')
ax2.tick_params(axis='x', rotation=45)  # 旋转x轴标签，避免重叠

# 第三个子图：利润率分布
profit_margin = (sales - expenses) / sales * 100
ax3.hist(profit_margin, bins=20, alpha=0.7, color='purple', edgecolor='black')
ax3.set_title('利润率分布')
ax3.set_xlabel('利润率 (%)')
ax3.set_ylabel('天数')
ax3.grid(True, alpha=0.3, axis='y')

# 调整布局
plt.tight_layout()

plt.show()
```
